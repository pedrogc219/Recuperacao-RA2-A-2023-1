##1.1 Resolução

$$f(x) = x^3 + 5x^2 - 7x + 9$$

Regra da soma:

$$ \frac{d}{dx}f(x) = \frac{d}{dx}x^3 + \frac{d}{dx}5x^2 - \frac{d}{dx}7x + \frac{d}{dx}9 $$

Regra da potência:

$$ \frac{d}{dx}f(x) = 3x^2 + 10x - 7 $$

```python
# gráfico equação 1
import numpy as np
import sympy as sy
import matplotlib.pyplot as plt

eixo_x = np.linspace(-10, 10, 200)

x = sy.symbols('x')
expr = x**3 + 5*x**2 - 7*x + 9
expr_derivada = sy.diff(expr, x)

f = sy.lambdify(x, expr)
df = sy.lambdify(x, expr_derivada)

plt.plot(eixo_x, f(eixo_x), label="$f(x)=x^3+5x^2-7x+9$") # curva da função
plt.plot(eixo_x, df(eixo_x), label="$f'(x)=3x^2+10x-7$") # curva da derivada
plt.grid(True)
plt.legend()
plt.show()
```

## 1.2
$$g(x) = e^x \cos(x)$$

Regra do produto sendo, $u=e^x$ e $v=cos(x)$:

$$ \frac{d}{dx}g(x) = e^x \frac{d}{dx}cos(x) + cos(x) \frac{d}{dx}e^x $$

$$ \frac{d}{dx}g(x) = -e^x sen(x) + cos(x) e^x $$

$$ \frac{d}{dx}g(x) = e^x(-sen(x)+cos(x)) $$

```python
# gráfico equaçào 2
import numpy as np
import sympy as sy
from math import e
import matplotlib.pyplot as plt

eixo_x = np.linspace(0, 11, 200)

x = sy.symbols('x')
expr = e**x * sy.cos(x)
expr_derivada = sy.diff(expr, x)

f = sy.lambdify(x, expr)
df = sy.lambdify(x, expr_derivada)

plt.plot(eixo_x, f(eixo_x), label="$g(x)=e^xcos(x)$") # curva da função
plt.plot(eixo_x, df(eixo_x), label="$g'(x)=e^x(-sen(x)+cos(x))$") # curva da derivada
plt.grid(True)
plt.legend()
plt.show()
```

Calcule a derivada da função: $j(x) = \sqrt{x} \ln(x)$ para $x=3$

Representando $\sqrt{x}$ como $x^{\frac{1}{2}}$ para simplificar o calculo:

$$ \frac{d}{dx}j(x) = x^{\frac{1}{2}}ln(x) $$

Regra do produto sendo, $u=x^{\frac{1}{2}}$ e $v=ln(x)$:

$$ \frac{d}{dx}j(x) = x^{\frac{1}{2}} \frac{d}{dx}ln(x) + ln(x) \frac{d}{dx}x^{\frac{1}{2}} $$

$$ \frac{d}{dx}j(x) = x^{\frac{1}{2}} \frac{1}{x} + ln(x) \frac{1}{2}x^{-\frac{1}{2}} $$

$$ \frac{d}{dx}j(x) = \sqrt{x} \frac{1}{x} + \frac{ln(x)}{2}\frac{1}{\sqrt{x}} $$

$$ \frac{d}{dx}j(x) = \frac{1}{\sqrt{x}} + \frac{ln(x)}{2\sqrt{x}} $$

Substituindo x:

$$ \frac{d}{dx}j(3) = \frac{1}{\sqrt{3}} + \frac{ln(3)}{2\sqrt{3}} $$

$$ \frac{d}{dx}j(3) = \frac{1}{1.73} + \frac{1.09}{2\times1.73} $$

$$ \frac{d}{dx}j(3) = \frac{1}{1.73} + \frac{1.09}{3.46} $$

Resultado é aproximadamente:

$$ \frac{d}{dx}j(3) = 0.89 $$


# Questão 3
Usando o python, numpy, encontre o máximo local de $$f(x)=x^2-98x+4$$ para $0 \leq x \leq 100$. 
 

```python
# Questão 3

# Função
def f(x):
    resultado = x**2 - 98*x + 4
    return resultado

# Achando intevalo onde estão o maximo e minimo
rangeMaximo = [f(0), f(1)]

for i in range(1, 101, 1):
    if f(i) > rangeMaximo[1]:
        rangeMaximo[0] = rangeMaximo[1]
        rangeMaximo[1] = i

print("rangeMaximo: ", rangeMaximo)

# Achando o maximo dentro do range
i = rangeMaximo[0]
maximo = f(i)
pontoMaximo = 0
while i < rangeMaximo[1]+0.01:
    if f(i) > maximo:
        maximo = f(i)
        pontoMaximo = i
    i += 0.01
print("Maximo no ponto", round(pontoMaximo,2), "valor", maximo)
print()
```
